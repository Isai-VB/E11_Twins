```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

outside = pd.read_csv('SensorsOutside.csv',skiprows=1)
outside2 = pd.read_csv('Sensors2.csv',skiprows=1)

```

```python
def plot_time_series(df, title):
    plt.figure(figsize=(10, 5))
    time_x_axis = np.arange(1, len(df) + 1)
    for col in df.columns:
        if col.lower() != 'current time':
            plt.plot(time_x_axis, df[col], label=col)
    
    plt.xlabel('Time (seconds)')
    plt.ylabel('Sensor Value')
    plt.yscale('log')
    plt.title(title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.tight_layout()
    plt.show()

# Plot all sensors in one figure using the full dataset
plot_time_series(outside, 'Sensors Outside Data vs Time')
plot_time_series(outside2, 'Sensors Outside2 Data vs Time')
```

```python
time_x_axis = np.arange(1, len(outside2) + 1)
sensor_columns = [col for col in outside2.columns if col.lower() != 'current time']

for sensor in sensor_columns:
    plt.figure(figsize=(10, 5))
    plt.plot(time_x_axis, outside2[sensor], label=f'Outside2 {sensor}')
    plt.plot(time_x_axis, outside[sensor], label=f'Outside {sensor}')
    plt.xlabel('Time (seconds)')
    plt.ylabel(sensor)
    plt.title(f'{sensor} vs Time (Outside2 vs Outside)')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
```

```python
sensor_columns = [col for col in outside2.columns if col.lower() != 'current time']

for sensor in sensor_columns:
    outside2_data = outside2[sensor].dropna()
    outside_data = outside[sensor].dropna()

    plt.figure(figsize=(10, 5))
    plt.hist(outside2_data, bins=50,  label=f'Outside2 {sensor}')
    plt.hist(outside_data, bins=70,  label=f'Outside {sensor}')
    plt.xlabel(sensor)
    plt.ylabel('Frequency')
    plt.title(f'{sensor} Distribution: Outside2 vs Outside')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

How do the data sets compare?

For the higher particles their data overlaps meaning that they similar in distribution. The other data we can see more spread meaning that there are differences in variability. 

```python
sensor_columns = [col for col in outside2.columns if col.lower() != 'current time']

for sensor in sensor_columns:
    outside2_data = outside2[sensor].dropna()
    outside_data = outside[sensor].dropna()

    mu_in = np.mean(outside2_data)
    mu_out = np.mean(outside_data)

    sigma_in = np.std(outside2_data, ddof=1)
    sigma_out = np.std(outside_data, ddof=1)

    n_in = len(outside2_data)
    n_out = len(outside_data)

    SE_in = sigma_in / np.sqrt(n_in)
    SE_out = sigma_out / np.sqrt(n_out)

    SE_delta = np.sqrt(SE_in**2 + SE_out**2)

    delta = abs(mu_in - mu_out)

    if delta > 3 * SE_delta:
        print(f"For sensor {sensor}, the means are separated by more than 3σ; significant difference.")
    else:
        print(f"For sensor {sensor}, the means are not separated by more than 3σ; no significant difference.")
```
