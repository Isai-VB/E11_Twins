```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_time_series(df, title):
    plt.figure(figsize=(10, 5))
    for col in df.columns:
        if col.lower() != 'current time':
            plt.plot(df['Current Time'], df[col], label=col)
    plt.xlabel('Time')
    plt.ylabel('Sensor Value')
    plt.yscale('log')
    plt.title(title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.tight_layout()
    plt.show()


# Plot all sensors in one figure
plot_time_series(outside, 'Sensors Outside Data vs Time')
plot_time_series(inside, 'Sensors Inside Data vs Time')
```

```python
inside['Current Time'] = pd.to_datetime(inside['Current Time'])
outside['Current Time'] = pd.to_datetime(outside['Current Time'])

sensor_columns = [col for col in inside.columns if col.lower() != 'current time']
for sensor in sensor_columns:
    plt.figure(figsize=(10, 5))
    plt.plot(inside['Current Time'], inside[sensor], label=f'Inside {sensor}')
    plt.plot(outside['Current Time'], outside[sensor], label=f'Outside {sensor}')
    plt.xlabel('Time')
    plt.ylabel(sensor)
    plt.title(f'{sensor} vs Time (Inside vs Outside)')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
```

```python
sensor_columns = [col for col in inside.columns if col.lower() != 'current time']

for sensor in sensor_columns:
    inside_data = inside[sensor].dropna()
    outside_data = outside[sensor].dropna()

    plt.figure(figsize=(10, 5))
    plt.hist(inside_data, bins=50,  label=f'Inside {sensor}')
    plt.hist(outside_data, bins=70,  label=f'Outside {sensor}')
    plt.xlabel(sensor)
    plt.ylabel('Frequency')
    plt.title(f'{sensor} Distribution: Inside vs Outside')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

How do the data sets compare?

For the higher particles their data overlaps meaning that they similar in distribution. The other data we can see more spread meaning that there are differences in variability. 

```python
sensor_columns = [col for col in inside.columns if col.lower() != 'current time']

for sensor in sensor_columns:
    inside_data = inside[sensor].dropna()
    outside_data = outside[sensor].dropna()

    mu_in = np.mean(inside_data)
    mu_out = np.mean(outside_data)

    sigma_in = np.std(inside_data, ddof=1)
    sigma_out = np.std(outside_data, ddof=1)

    n_in = len(inside_data)
    n_out = len(outside_data)

    SE_in = sigma_in / np.sqrt(n_in)
    SE_out = sigma_out / np.sqrt(n_out)

    SE_delta = np.sqrt(SE_in**2 + SE_out**2)

    delta = abs(mu_in - mu_out)

    if delta > 3 * SE_delta:
        print(f"For sensor {sensor}, the means are separated by more than 3σ; statistically significant difference.")
    else:
        print(f"For sensor {sensor}, the means are not separated by more than 3σ; no significant difference.")
```
